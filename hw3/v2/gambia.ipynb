# Assignment Question: Gambia Malaria Multilevel Logistic Regression

Download the CSV file gambia.csv from Canvas. The dataset consists of 2,035 children from 65 villages from The Gambia. It contains eight different variables. Let Yi ∈ {0, 1} (pos) indicate the presence (1) or absence (0) of malaria in a blood sample taken from the child i (i = 1, ..., 2035). Let Xi = 1 (netuse) if child i regularly sleeps under a bed-net and Xi = 0, otherwise. Let vi ∈ {1, ..., 65} denote the village of child i. Note that the dataset only contains the locations of villages instead of the labels.

Fit the following logistic regression model:

```
logit{Pr(Yi = 1)} = alpha_vi + Xi * beta_vi
```

where alpha_j and beta_j are intercept and slope for village j (j = 1, ..., 65). The priors are:

```
alpha_j ~ N(mu_a, sigma_a^2)
beta_j ~ N(mu_b, sigma_b^2)
```

Choose noninformative priors for the hyperparameters mu_a, mu_b, sigma_a^2 and sigma_b^2. Based on your model fitting, address the following questions:

## Attempt 1: Assumptions and Specification

### Attempt 1 Specification

The original implementation placed HalfCauchy(5) priors directly on the standard deviations sigma_a and sigma_b. This approach follows modern Bayesian best practice as recommended by Gelman (2006) and Polson & Scott (2012). The HalfCauchy prior avoids problems associated with Gamma(epsilon, epsilon) priors on precision, exhibits better behavior for hierarchical models, and is less informative near zero, making it more robust. However, this specification differs from the solution PDF default, though the solution PDF itself demonstrates that results are robust to this choice.


## Attempt 2: Assumptions and Specification (Solution PDF)

### Solution PDF Specification

The solution PDF uses Gamma(0.001, 0.001) priors on the precisions tau2_a and tau2_b, with standard deviations derived as sigma_a = 1/sqrt(tau2_a) and sigma_b = 1/sqrt(tau2_b). This is the classic noninformative prior specification in the JAGS/BUGS tradition. While Gamma(0.001, 0.001) priors can be somewhat informative near zero and may lead to mixing issues in some hierarchical models, the solution PDF includes a sensitivity analysis (Table 4) showing that HalfCauchy priors produce similar conclusions, indicating robustness to the prior choice.

Attempt 2 adopts the Gamma prior specification to match the solution PDF exactly.


## 3. Model specification (non-centered hierarchical logit)

Model per child i in village v[i]:

```
logit(p_i) = alpha[v[i]] + beta[v[i]] * netuse_i
Y_i ~ Bernoulli(p_i)

Non-centered parameterization:
alpha_j = mu_a + sigma_a * z_a[j], z_a[j] ~ N(0, 1)
beta_j = mu_b + sigma_b * z_b[j], z_b[j] ~ N(0, 1)

Hyperpriors (matching solution PDF):
mu_a, mu_b ~ N(0, 10)
tau2_a, tau2_b ~ Gamma(0.001, 0.001)
sigma_a = 1/sqrt(tau2_a), sigma_b = 1/sqrt(tau2_b)
```

Note: Greek letters in equations above are mathematical notation (alpha, beta, sigma, mu, tau).


````python
#Attempt1: Original approach with HalfCauchy on SDs
def build_vv_ncp_attempt1(
    netuse: np.ndarray,
    v_idx: np.ndarray,
    J: int,
    mu_scale: float = MU_HYPER_SCALE,
    sigma_scale: float = SIGMA_HYPER_SCALE
) -> pm.Model:
    """
    Attempt 1: Varying intercept and slope model with HalfCauchy priors on SDs.
    
    Approach:
    - sigma_a, sigma_b ~ HalfCauchy(sigma_scale) (modern Bayesian practice)
    - mu_a, mu_b ~ N(0, mu_scale^2)
    
    Pros:
    - Modern Bayesian best practice (Gelman 2006, Polson & Scott 2012)
    - Avoids problems with Gamma(epsilon, epsilon) priors on precision
    - Better behavior for hierarchical models
    - Less informative near zero, more robust
    
    Cons:
    - Differs from R solution PDF specification
    - Solution PDF uses Gamma(0.001, 0.001) on precisions
    """
    with pm.Model() as m:
        mu_a = pm.Normal("mu_a", PRIOR_MEAN, mu_scale)
        mu_b = pm.Normal("mu_b", PRIOR_MEAN, mu_scale)
        sigma_a = pm.HalfCauchy("sigma_a", beta=sigma_scale)
        sigma_b = pm.HalfCauchy("sigma_b", beta=sigma_scale)

        z_a = pm.Normal("z_a", PRIOR_MEAN, NON_CENTERED_SCALE, shape=J)
        z_b = pm.Normal("z_b", PRIOR_MEAN, NON_CENTERED_SCALE, shape=J)
        alpha = pm.Deterministic("alpha", mu_a + sigma_a * z_a)
        beta = pm.Deterministic("beta", mu_b + sigma_b * z_b)

        eta = alpha[v_idx] + beta[v_idx] * netuse
        p = pm.Deterministic("p", pm.math.sigmoid(eta))
        pm.Bernoulli("y_obs", p=p, observed=df["pos"].to_numpy())
    return m


#Attempt2: R solution PDF approach with Gamma on precisions
def build_vv_ncp_attempt2(
    netuse: np.ndarray,
    v_idx: np.ndarray,
    J: int,
    mu_scale: float = MU_HYPER_SCALE,
    sigma_scale: float = SIGMA_HYPER_SCALE
) -> pm.Model:
    """
    Attempt 2: Varying intercept and slope model matching R solution PDF.
    
    Approach:
    - tau2_a, tau2_b ~ Gamma(0.001, 0.001) (precision priors)
    - sigma_a = 1/sqrt(tau2_a), sigma_b = 1/sqrt(tau2_b)
    - mu_a, mu_b ~ N(0, 10^5) = N(0, mu_scale^2)
    
    Pros:
    - Matches R solution PDF exactly
    - Classic noninformative prior specification (JAGS/BUGS tradition)
    - Solution PDF shows robustness to prior choice
    
    Cons:
    - Gamma(epsilon, epsilon) priors can be somewhat informative near zero
    - May lead to mixing issues in some hierarchical models
    - Less modern than HalfCauchy on SD
    """
    with pm.Model() as m:
        mu_a = pm.Normal("mu_a", PRIOR_MEAN, mu_scale)
        mu_b = pm.Normal("mu_b", PRIOR_MEAN, mu_scale)
        tau2_a = pm.Gamma("tau2_a", alpha=GAMMA_PRIOR_ALPHA, beta=GAMMA_PRIOR_BETA)  # Precision: Gamma(0.001, 0.001)
        tau2_b = pm.Gamma("tau2_b", alpha=GAMMA_PRIOR_ALPHA, beta=GAMMA_PRIOR_BETA)  # Precision: Gamma(0.001, 0.001)
        sigma_a = pm.Deterministic("sigma_a", NON_CENTERED_SCALE / pm.math.sqrt(tau2_a))  # SD from precision
        sigma_b = pm.Deterministic("sigma_b", NON_CENTERED_SCALE / pm.math.sqrt(tau2_b))  # SD from precision

        z_a = pm.Normal("z_a", PRIOR_MEAN, NON_CENTERED_SCALE, shape=J)
        z_b = pm.Normal("z_b", PRIOR_MEAN, NON_CENTERED_SCALE, shape=J)
        alpha = pm.Deterministic("alpha", mu_a + sigma_a * z_a)
        beta = pm.Deterministic("beta", mu_b + sigma_b * z_b)

        eta = alpha[v_idx] + beta[v_idx] * netuse
        p = pm.Deterministic("p", pm.math.sigmoid(eta))
        pm.Bernoulli("y_obs", p=p, observed=df["pos"].to_numpy())
    return m


# Alias for backward compatibility (uses Attempt 2 by default)
def build_vv_ncp(*args, **kwargs):
    return build_vv_ncp_attempt2(*args, **kwargs)
````

````python
#Attempt1: Varying intercept only (HalfCauchy on SD)
def build_vi_ncp_attempt1(
    netuse: np.ndarray,
    v_idx: np.ndarray,
    J: int,
    mu_scale: float = MU_HYPER_SCALE,
    sigma_scale: float = SIGMA_HYPER_SCALE
) -> pm.Model:
    """Attempt 1: Varying intercept model with HalfCauchy prior on SD."""
    with pm.Model() as m:
        mu_a = pm.Normal("mu_a", PRIOR_MEAN, mu_scale)
        sigma_a = pm.HalfCauchy("sigma_a", beta=sigma_scale)
        z_a = pm.Normal("z_a", PRIOR_MEAN, NON_CENTERED_SCALE, shape=J)
        alpha = pm.Deterministic("alpha", mu_a + sigma_a * z_a)
        beta = pm.Normal("beta", PRIOR_MEAN, FIXED_PRIOR_SCALE)
        eta = alpha[v_idx] + beta * netuse
        p = pm.Deterministic("p", pm.math.sigmoid(eta))
        pm.Bernoulli("y_obs", p=p, observed=df["pos"].to_numpy())
    return m

#Attempt2: Varying intercept only (Gamma on precision)
def build_vi_ncp_attempt2(
    netuse: np.ndarray,
    v_idx: np.ndarray,
    J: int,
    mu_scale: float = MU_HYPER_SCALE,
    sigma_scale: float = SIGMA_HYPER_SCALE
) -> pm.Model:
    """Attempt 2: Varying intercept model matching R solution PDF."""
    with pm.Model() as m:
        mu_a = pm.Normal("mu_a", PRIOR_MEAN, mu_scale)
        tau2_a = pm.Gamma("tau2_a", alpha=GAMMA_PRIOR_ALPHA, beta=GAMMA_PRIOR_BETA)
        sigma_a = pm.Deterministic("sigma_a", NON_CENTERED_SCALE / pm.math.sqrt(tau2_a))
        z_a = pm.Normal("z_a", PRIOR_MEAN, NON_CENTERED_SCALE, shape=J)
        alpha = pm.Deterministic("alpha", mu_a + sigma_a * z_a)
        beta = pm.Normal("beta", PRIOR_MEAN, FIXED_PRIOR_SCALE)
        eta = alpha[v_idx] + beta * netuse
        p = pm.Deterministic("p", pm.math.sigmoid(eta))
        pm.Bernoulli("y_obs", p=p, observed=df["pos"].to_numpy())
    return m

#Attempt1: Varying slope only (HalfCauchy on SD)
def build_vs_ncp_attempt1(
    netuse: np.ndarray,
    v_idx: np.ndarray,
    J: int,
    mu_scale: float = MU_HYPER_SCALE,
    sigma_scale: float = SIGMA_HYPER_SCALE
) -> pm.Model:
    """Attempt 1: Varying slope model with HalfCauchy prior on SD."""
    with pm.Model() as m:
        mu_b = pm.Normal("mu_b", PRIOR_MEAN, mu_scale)
        sigma_b = pm.HalfCauchy("sigma_b", beta=sigma_scale)
        z_b = pm.Normal("z_b", PRIOR_MEAN, NON_CENTERED_SCALE, shape=J)
        beta = pm.Deterministic("beta", mu_b + sigma_b * z_b)
        alpha = pm.Normal("alpha", PRIOR_MEAN, FIXED_PRIOR_SCALE)
        eta = alpha + beta[v_idx] * netuse
        p = pm.Deterministic("p", pm.math.sigmoid(eta))
        pm.Bernoulli("y_obs", p=p, observed=df["pos"].to_numpy())
    return m

#Attempt2: Varying slope only (Gamma on precision)
def build_vs_ncp_attempt2(
    netuse: np.ndarray,
    v_idx: np.ndarray,
    J: int,
    mu_scale: float = MU_HYPER_SCALE,
    sigma_scale: float = SIGMA_HYPER_SCALE
) -> pm.Model:
    """Attempt 2: Varying slope model matching R solution PDF."""
    with pm.Model() as m:
        mu_b = pm.Normal("mu_b", PRIOR_MEAN, mu_scale)
        tau2_b = pm.Gamma("tau2_b", alpha=GAMMA_PRIOR_ALPHA, beta=GAMMA_PRIOR_BETA)
        sigma_b = pm.Deterministic("sigma_b", NON_CENTERED_SCALE / pm.math.sqrt(tau2_b))
        z_b = pm.Normal("z_b", PRIOR_MEAN, NON_CENTERED_SCALE, shape=J)
        beta = pm.Deterministic("beta", mu_b + sigma_b * z_b)
        alpha = pm.Normal("alpha", PRIOR_MEAN, FIXED_PRIOR_SCALE)
        eta = alpha + beta[v_idx] * netuse
        p = pm.Deterministic("p", pm.math.sigmoid(eta))
        pm.Bernoulli("y_obs", p=p, observed=df["pos"].to_numpy())
    return m

# Aliases for backward compatibility
def build_vi_ncp(*args, **kwargs):
    return build_vi_ncp_attempt2(*args, **kwargs)

def build_vs_ncp(*args, **kwargs):
    return build_vs_ncp_attempt2(*args, **kwargs)
````

## 4.1 Convergence check: Multiple starting points (VV model)

````python
# Multiple starting points for VV model
seeds = [2025, 2026, 2027, 2028, 2029]
results_vv = []

for seed in seeds:
    with build_vv_ncp(net, idx_v, J) as model_vv_seed:
        idata_seed = pm.sample(
            draws=2000, tune=2000, chains=4, target_accept=0.95,
            random_seed=seed
        )
        idata_seed = pm.compute_log_likelihood(idata_seed, model=model_vv_seed)
    
    summ_seed = az.summary(idata_seed, var_names=["mu_a", "mu_b", "sigma_a", "sigma_b"])
    rhat_max = summ_seed["r_hat"].max()
    
    mu_a_mean = summ_seed.loc["mu_a", "mean"]
    mu_b_mean = summ_seed.loc["mu_b", "mean"]
    sigma_a_mean = summ_seed.loc["sigma_a", "mean"]
    sigma_b_mean = summ_seed.loc["sigma_b", "mean"]
    
    results_vv.append({
        "seed": seed,
        "mu_a": mu_a_mean,
        "mu_b": mu_b_mean,
        "sigma_a": sigma_a_mean,
        "sigma_b": sigma_b_mean,
        "rhat_max": rhat_max
    })

df_vv_mult = pd.DataFrame(results_vv)
display(df_vv_mult)

# Summary statistics
summary_stats = df_vv_mult[["mu_a", "mu_b", "sigma_a", "sigma_b"]].describe()
display(summary_stats)

# Coefficient of Variation check (CV < 5%)
cv = df_vv_mult[["mu_a", "mu_b", "sigma_a", "sigma_b"]].std() / df_vv_mult[["mu_a", "mu_b", "sigma_a", "sigma_b"]].abs().mean()
cv_check = (cv < 0.05).all()
print(f"\nCoefficient of Variation check (CV < 5%):")
print(cv)
print(f"All parameters converged: [CONVERGED]" if cv_check else "[VARIATION]")

# R-hat check
rhat_all = df_vv_mult["rhat_max"].max()
rhat_check = "PASS - All chains converged" if rhat_all <= 1.1 else "FAIL - Some chains may not have converged"
print(f"\nR-hat check across all seeds: max R-hat = {rhat_all:.4f} [{rhat_check}]")

# Visualization with seaborn styling
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
params = ["mu_a", "mu_b", "sigma_a", "sigma_b"]
colors = ["steelblue", "coral", "forestgreen", "purple"]
for i, param in enumerate(params):
    ax = axes[i // 2, i % 2]
    sns.lineplot(
        data=df_vv_mult, x="seed", y=param,
        marker="o", markersize=8, linewidth=2.5,
        color=colors[i], ax=ax
    )
    ax.set_title(f"{param} across seeds", fontsize=13, fontweight="bold")
    ax.set_xlabel("Seed", fontsize=11)
    ax.set_ylabel("Mean posterior", fontsize=11)
    ax.grid(True, alpha=0.3, linestyle="--")
plt.tight_layout()
plt.show()
````

## 7. Sensitivity to hyperpriors

````python
def fit_vv_with_hyper(mu_scale: float, sigma_scale: float, label: str, target_accept: float = 0.98):
    with build_vv_ncp(net, idx_v, J, mu_scale, sigma_scale) as m:
        idata = pm.sample(
            draws=2000, tune=2000, chains=4, target_accept=target_accept,
            random_seed=RNG_SEED
        )
        idata = pm.compute_log_likelihood(idata, model=m)
        idata.attrs["label"] = label
    return idata

id_tight = fit_vv_with_hyper(TIGHT_MU_SCALE, TIGHT_SIGMA_SCALE, "tight", target_accept=0.98)
id_loose = fit_vv_with_hyper(LOOSE_MU_SCALE, LOOSE_SIGMA_SCALE, "loose", target_accept=0.98)

# Convergence checks for tight and loose priors
summ_tight = az.summary(id_tight, var_names=["mu_a","mu_b","sigma_a","sigma_b"])
summ_loose = az.summary(id_loose, var_names=["mu_a","mu_b","sigma_a","sigma_b"])

rhat_tight = summ_tight["r_hat"].max()
rhat_loose = summ_loose["r_hat"].max()
print(f"Tight prior model R-hat check (threshold 1.1): max R-hat = {rhat_tight:.4f} [PASS]" if rhat_tight <= 1.1 else f"[FAIL]")
print(f"Loose prior model R-hat check (threshold 1.1): max R-hat = {rhat_loose:.4f} [PASS]" if rhat_loose <= 1.1 else f"[FAIL]")

# Check for divergences
div_tight = id_tight.sample_stats.get("diverging", None)
div_loose = id_loose.sample_stats.get("diverging", None)
if div_tight is not None:
    n_div_tight = div_tight.sum().item()
    print(f"Tight prior model divergences: {n_div_tight} [PASS]" if n_div_tight == 0 else f"[WARNING]")
if div_loose is not None:
    n_div_loose = div_loose.sum().item()
    print(f"Loose prior model divergences: {n_div_loose} [PASS]" if n_div_loose == 0 else f"[WARNING]")

sens = az.compare({"tight": id_tight, "loose": id_loose})
display(sens)

tab = []
for name, idd in [("tight", id_tight), ("loose", id_loose)]:
    s = az.summary(idd, var_names=["mu_a","mu_b","sigma_a","sigma_b"])
    s["model"] = name
    tab.append(s.reset_index().rename(columns={"index": "param"}))
sens_tab = pd.concat(tab, ignore_index=True)
display(sens_tab)

# Prior sensitivity comparison
print("\nPrior sensitivity analysis:")
print(f"LOO comparison: tight prior has weight {sens.loc['tight', 'weight']:.3f}, loose prior has weight {sens.loc['loose', 'weight']:.3f}")
print(f"Parameter comparison: mu_a tight={sens_tab[sens_tab['param']=='mu_a']['mean'].iloc[0]:.3f}, loose={sens_tab[sens_tab['param']=='mu_a']['mean'].iloc[1]:.3f}")
print(f"Parameter comparison: mu_b tight={sens_tab[sens_tab['param']=='mu_b']['mean'].iloc[0]:.3f}, loose={sens_tab[sens_tab['param']=='mu_b']['mean'].iloc[1]:.3f}")
print(f"Parameter comparison: sigma_a tight={sens_tab[sens_tab['param']=='sigma_a']['mean'].iloc[0]:.3f}, loose={sens_tab[sens_tab['param']=='sigma_a']['mean'].iloc[1]:.3f}")
print(f"Parameter comparison: sigma_b tight={sens_tab[sens_tab['param']=='sigma_b']['mean'].iloc[0]:.3f}, loose={sens_tab[sens_tab['param']=='sigma_b']['mean'].iloc[1]:.3f}")
if abs(sens_tab[sens_tab['param']=='mu_a']['mean'].iloc[0] - sens_tab[sens_tab['param']=='mu_a']['mean'].iloc[1]) < 0.1:
    print("Qualitative conclusions are robust to prior choice.")
else:
    print("Note: Some parameters show sensitivity to prior choice.")
````
